```python
missing_vals = df.isna().sum().sort_values() / len(df)
missing_vals = missing_vals[missing_vals > 0]   #filter
# missing_vals.shape

#divide missing columns into two groups
missing_vals_first = missing_vals.head(len(missing_vals)//2 + 1)
missing_vals_last = missing_vals.tail(len(missing_vals)//2)
```

```python
missing_vals_first
```

```python
plt.figure(figsize=(20, 8))
plt.title('Missing Value Percentage', fontdict=dict(fontsize=30))
missing_vals_first.plot(kind = 'bar', color='xkcd:hunter green')
plt.grid(True, which='both', linestyle='--')
plt.xlabel('Features')
plt.ylabel('Missing Percentage')

for col, data in enumerate(missing_vals_first):
    plt.text(x=col - 0.2, y = data + 0.01, s=f"{data * 100:.{2}f}%" , fontdict=dict(fontsize=15))

plt.figure(figsize=(20, 8))
missing_vals_last.plot(kind = 'bar', color='xkcd:hunter green')
plt.grid(True, which='both', linestyle='--')
for col, data in enumerate(missing_vals_last):
    plt.text(x=col - 0.2, y = data + 0.01, s=f"{data * 100:.{2}f}%" , fontdict=dict(fontsize=15))
plt.xlabel('Features')
plt.ylabel('CountMissing')
plt.show()
```

```python
df.drop(['Fence', 'Alley', 'Misc Feature', 'Pool QC'], axis=1, inplace=True)
df.head()
```

```python
#remove duplicate
df = df.drop_duplicates()
df.shape
```

# Encode

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for col in df.columns:
    if df[col].dtype == 'O':
        series = df[col]
        label_encoder = LabelEncoder()
        df[col] = pd.Series(label_encoder.fit_transform(series[series.notnull()]), index=series[series.notnull()].index)
```

```python
missing = df.isna().sum()
missing[missing>0]
```

# Imputation

```python
#import libraries
from sklearn.impute import KNNImputer
```

```python
impute = KNNImputer(n_neighbors=10, weights='distance')
df_imputed = pd.DataFrame(impute.fit_transform(df), columns=df.columns)
df_imputed.isna().sum()
```

```python
for col in df_imputed.columns:
    df_imputed[col] = df_imputed[col].astype('float32')
```

```python
from sklearn.preprocessing import MinMaxScaler
min_max = MinMaxScaler().fit_transform(df_imputed)
min_max
```
