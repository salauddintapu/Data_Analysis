# Correlation

Plot top 15 correlated features

```python
plt.figure(figsize=(10,10))
corr = df.corr()
top_corr_cols = corr.abs().nlargest(15, columns=corr.columns).index
top_corr = corr.loc[top_corr_cols, top_corr_cols]
fig, axes = plt.subplots(figsize=(15, 15))
mask = np.zeros_like(top_corr)
mask[np.triu_indices_from(mask)] = True
sns.heatmap(top_corr, mask=mask, linewidths=.5, cmap=sns.cubehelix_palette(as_cmap=True), annot=True)
plt.show()
```

Find correlation above a specific threshold (0.5|)

```python
hous_num = df.select_dtypes(include = ['float64', 'int64'])
hous_num_corr = hous_num.corr()['SalePrice'][:-1] # -1 means that the latest row is SalePrice
top_features = hous_num_corr[abs(hous_num_corr) > 0.5].sort_values(ascending=True) #displays pearsons correlation coefficient greater than 0.5
print("There is {} strongly correlated values with SalePrice:\n{}".format(len(top_features), top_features))
```

```python
#plot correlation with target
for i in range(0, len(hous_num.columns), 5):
    sns.pairplot(data=hous_num,
                x_vars=hous_num.columns[i:i+5],
                y_vars=['SalePrice'])
```

# Handle Duplicate and Missing Values

```python
#find duplicates
df.duplicated().sum()
```

```python
#remove duplicaets
df= df.drop_duplicates()
```

```python
#plot missing values by columns

missing_by_col =  df.isna().sum().sort_values() / len(df)
missing_by_col = missing_by_col[missing_by_col > 0]
missing_by_col_first = missing_by_col.head(len(missing_by_col)//2+1)
missing_by_col_second = missing_by_col.tail(len(missing_by_col)//2)

plt.figure(figsize=(20, 8))
plt.title('Missing values count', fontdict=dict(fontsize=30))
missing_by_col_first.plot(kind = 'bar', color='xkcd:hunter green')
plt.grid(True, which='both', linestyle='--')
plt.xlabel('Features')
plt.ylabel('CountMissing');

for index,data in enumerate(missing_by_col_first):
    plt.text(x=index - 0.2, y = data + 0.001, s=f"{data * 100:.{2}f}%" , fontdict=dict(fontsize=15))
plt.show();

plt.figure(figsize=(20, 8))
missing_by_col_second.plot(kind = 'bar', color='xkcd:hunter green')
plt.grid(True, which='both', linestyle='--')
for index,data in enumerate(missing_by_col_second):
    plt.text(x=index - 0.2, y = data + 0.01, s=f"{data * 100:.{2}f}%" , fontdict=dict(fontsize=15))
plt.xlabel('Features')
plt.ylabel('CountMissing')
```
